# Combining DoS (RSS) and TA (API) data:

## 1. Importing and loading the respective data sets

```python
import pandas as pd
```

```python
ta = pd.read_csv('./data/ta_api.csv', index_col=0)
ta.head()
```

```python
dos = pd.read_csv('./data/dos_rss.csv', index_col=0)
dos.head()
```

## 2. Merging the data sets

```python
print(dos.values.shape, ta.values.shape)
```

```python
common = ta.merge(dos, on='Name', suffixes=('_TA', '_DOS'))
```

```python
common.shape
```

```python
common.head()
```

```python
common = common.drop('ISO_DOS', axis=1)
```

```python
common.head()
```

```python
common.loc[common['Name']=='China']
```
